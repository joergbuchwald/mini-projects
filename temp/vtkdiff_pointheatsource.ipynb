```python
import vtuIO
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.rcParams['font.size'] = 22
#plt.rcParams['lines.linewidth']= 2.0
plt.rcParams['legend.frameon']=True
plt.rcParams['figure.figsize'] = (20, 15)
plt.rcParams['font.family'] = 'serif'
plt.rcParams['legend.fontsize']=22
plt.rcParams["mathtext.fontset"] = "dejavuserif"
plt.rcParams['font.size'] = 22
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.left'] = True
plt.rcParams['axes.spines.bottom'] = True
plt.rcParams['axes.axisbelow'] = True
```

```python
pointheatsource_bugfix = vtuIO.PVDIO(".", "square_1e0_pcs_0_bugfix.pvd", dim=2)
```

```python
pointheatsource_old = vtuIO.PVDIO(".", "square_1e0_pcs_0_old.pvd", dim=2)
```

```python
pt = {'pt0': (0.1,0.1, 0.0)}
```

```python
temp_bugfix = pointheatsource_bugfix.readTimeSeries("temperature_interpolated", pts=pt)
```

```python
press_bugfix = pointheatsource_bugfix.readTimeSeries("pressure_interpolated", pts=pt)
```

```python
temp_old = pointheatsource_old.readTimeSeries("temperature_interpolated", pts=pt)
press_old = pointheatsource_old.readTimeSeries("pressure_interpolated", pts=pt)
```

# temperature mixed Taylor-Hood mesh

```python
plt.plot(pointheatsource_old.ts_files["ts"], temp_old['pt0'], label="old")
plt.plot(pointheatsource_bugfix.ts_files["ts"], temp_bugfix['pt0'], label="bugfix")
plt.title("temperature")
plt.ylabel("T/K")
plt.xlabel("t/s")
plt.legend()
```

```python
plt.plot(pointheatsource_old.ts_files["ts"], np.array(temp_old['pt0'])-np.array(temp_bugfix['pt0']))
plt.title("temperature error")
plt.ylabel("$\\Delta T$/K")
plt.xlabel("t/s")
```

# pressure mixed Taylor-Hood mesh

```python
plt.plot(pointheatsource_old.ts_files["ts"], press_old['pt0'], label="old")
plt.plot(pointheatsource_bugfix.ts_files["ts"], press_bugfix['pt0'], label="bugfix")
plt.title("pressure")
plt.ylabel("p/Pa")
plt.xlabel("t/s")
plt.legend()
```

```python
plt.plot(pointheatsource_old.ts_files["ts"], np.array(press_old['pt0'])-np.array(press_bugfix['pt0']))
plt.title("pressure error")
plt.ylabel("$\\Delta p$/Pa")
plt.xlabel("t/s")
```
