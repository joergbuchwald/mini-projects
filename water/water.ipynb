```python
import numpy as np
import matplotlib.pyplot as plt

from water import water
```

```python
f_eng_toolbox = water.WATER(rho=0,a=0)
f_poly2 = water.WATER(rho=2,a=1)
f_exp = water.WATER(rho=3)
f_poly3 = water.WATER(rho=4,a=2)
f_IAPWS = water.WATER(rho=5)
```

```python
T = f_eng_toolbox.example_T
```

```python
plt.title("exp. density model vs temperature")
plt.plot(T, f_eng_toolbox.rho.value(T), label="eng toolbox")
plt.plot(T, f_poly2.rho.value(T,1e6), label=f_poly2.rho.exprtk_value())
plt.plot(T, f_poly3.rho.value(T,1e6), label=f_poly3.rho.exprtk_value())
plt.plot(T, f_exp.rho.value(T,1e6), label=f_exp.rho.exprtk_value())
plt.plot(T, f_IAPWS.rho.value(T,1e6), label="IAPWS")
plt.legend()
plt.xlabel("T / K")
plt.ylabel("rho / kg m$^-3$")
plt.show()

```

```python
plt.title("derivative of density model with respect to temperature")
plt.plot(T, f_eng_toolbox.rho.dvalue(T), label="eng toolbox")
plt.plot(T, f_poly2.rho.dvalue(T,1e6, variable="temperature"), label=f_poly2.rho.exprtk_dvalue())
plt.plot(T, f_poly3.rho.dvalue(T,1e6, variable="temperature"), label=f_poly3.rho.exprtk_dvalue())
plt.plot(T, f_exp.rho.dvalue(T,1e6, variable="temperature"), label=f_exp.rho.exprtk_dvalue())
plt.plot(T, f_IAPWS.rho.dvalue(T,1e6, variable="temperature"), label="IAPWS")
plt.legend()
plt.xlabel("T / K")
plt.ylabel("Drho / kg m$^{-3}$ K$^{-1}$")
plt.show()
```

```python
plt.title("expansivity vs temperature")
plt.plot(T, f_eng_toolbox.a.value(f_eng_toolbox.example_T), label="eng toolbox")
plt.plot(T, f_poly2.a.value(T), label=f_poly2.a.exprtk_value())
plt.plot(T, f_poly3.a.value(T), label=f_poly3.a.exprtk_value())
plt.legend()
plt.xlabel("T / K")
plt.ylabel("a / K$^{-1}$")
plt.show()
```
